# Two qubit gates: CNOT

```python
# Create a Quantum Circuit acting on a quantum register of two qubit
cr = ClassicalRegister(2)
qr = QuantumRegister(2)
circ = QuantumCircuit(qr,cr)

# Add a CNOT gate between both the qubits. 
circ.cx(qr[0], qr[1])

circ.measure(qr, cr)
circ.draw(output='mpl')
```
